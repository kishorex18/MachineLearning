```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn import tree
```

```python
df=pd.read_csv("titanic.csv")
df=df.drop(["PassengerId","Name","SibSp","Parch","Cabin","Embarked","Ticket"],axis=1)
df.head()
```

```python
le=LabelEncoder()
df["Sex"]=le.fit_transform(df["Sex"])
df.head()
```

```python
X=df.drop("Survived",axis=1)
Y=df["Survived"]
X_train,X_test,Y_train,Y_test=train_test_split(X,Y)
```

```python
model=tree.DecisionTreeClassifier(max_depth=4,random_state=42)
```

```python
model.fit(X_train,Y_train)
```

```python
model.predict(X_test)
```

```python
model.score(X_test,Y_test)
```

```python
from sklearn.tree import plot_tree
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(30, 18))  # Adjust figure size
plot_tree(
    model, 
    feature_names=X.columns,   # Use DataFrame column names as feature names
    filled=True,              # Fill with colors for clarity
    rounded=True,             # Use rounded edges
    fontsize=10               # Font size for text
)
plt.axis('off')  # Turn off axis for a cleaner plot
```
