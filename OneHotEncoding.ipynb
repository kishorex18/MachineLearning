```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df=pd.read_csv("carprices.csv")
df
```

```python
plt.scatter(df["Car Model"],df["Sell Price($)"])
```

```python
plt.scatter(df["Age(yrs)"],df["Sell Price($)"])
```

```python
plt.scatter(df["Mileage"],df["Sell Price($)"])
```

```python
dummy=pd.get_dummies(df["Car Model"]).astype(int)
dummy
```

```python
dummy=dummy.drop("Audi A5",axis=1)
```

```python
df=df.drop("Car Model",axis=1)
```

```python
merged=pd.concat([df,dummy],axis=1)
merged
```

```python
X=merged.drop("Sell Price($)",axis=1)
X
```

```python
Y=df["Sell Price($)"]
Y
```

```python
reg=linear_model.LinearRegression()
```

```python
reg.fit(X,Y)
```

```python
X
```

```python
reg.predict([[45000,4,0,1]])
```

```python
reg.predict([[86000,7,1,0]])
```

```python
reg.score(X,Y)
```
