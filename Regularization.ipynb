```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression,Lasso,Ridge
```

```python
df=pd.read_csv("HousingDataset.csv")
```

```python
df.head()
```

```python
df.head()
```

```python

df.head()
```

```python
df.isnull().sum()
```

```python
df=df.fillna({"Bedroom2":0,"Bathroom":0,"Car":0})
df["Landsize"]=df["Landsize"].fillna(df["Landsize"].mean())
```

```python
df["Distance"]=df["Distance"].fillna(df["Distance"].mean())
df["YearBuilt"]=df["YearBuilt"].fillna(df["YearBuilt"].mean())
df["Lattitude"]=df["Lattitude"].fillna(df["Lattitude"].mean())
df["Longtitude"]=df["Longtitude"].fillna(df["Longtitude"].mean())
df["Propertycount"]=df["Propertycount"].fillna(df["Propertycount"].mean())
```

```python
df.drop("Postcode",inplace=True,axis=1)
```

```python
df.head()
df.drop(["Suburb","Address","Type","Method","SellerG","Date","BuildingArea","YearBuilt","CouncilArea","Regionname"],axis=1,inplace=True)
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace=True)
```

```python
X=df.drop("Price",axis=1)
Y=df.Price
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2)
df
```

```python
model1=LinearRegression()
```

```python
model1.fit(X_train,Y_train)
```

```python
model1.predict(X_test)
```

```python
model1.score(X_test,Y_test)
```

```python
model1.score(X_train,Y_train)
```

```python
model2=Lasso()
```

```python
model2.fit(X_train,Y_train)
```

```python
model2.predict(X_test)
```

```python
model2.score(X_test,Y_test)
```

```python
model2.score(X_train,Y_train)
```

```python
model3=Ridge()
```

```python
model3.fit(X_train,Y_train)
```

```python
model3.predict(X_test)
```

```python
model3.score(X_test,Y_test)
```
