```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler,LabelEncoder
from sklearn.model_selection import train_test_split
```

```python
df=pd.read_csv("heart.csv")
df
```

```python
dummy_sex=pd.get_dummies(df["Sex"]).astype(int)
df.drop(["Sex"],axis=1,inplace=True)
df=pd.concat([df,dummy_sex],axis=1)
df
```

```python
Label_cp=LabelEncoder()
Label_re=LabelEncoder()
Label_ea=LabelEncoder()
Label_ss=LabelEncoder()
```

```python
df["ChestPainType"]=Label_cp.fit_transform(df["ChestPainType"])
```

```python
df["RestingECG"]=Label_re.fit_transform(df["RestingECG"])
df["ExerciseAngina"]=Label_ea.fit_transform(df["ExerciseAngina"])
df["ST_Slope"]=Label_ss.fit_transform(df["ST_Slope"])
df.head()
```

```python
SS=StandardScaler()
```

```python
Y=df.HeartDisease
X=df.drop("HeartDisease",axis=1)
X_scaled=pd.DataFrame(SS.fit_transform(X),columns=X.columns)
X_scaled.head()
```

```python
X_scaled=pd.concat([X_scaled,Y],axis=1)
mask = (X_scaled.abs() <= 3).all(axis=1)
X_scaled=X_scaled[mask]
X_scaled
```

```python
X_scaled.describe()
Y_scaled=X_scaled["HeartDisease"]
X_scaled.drop("HeartDisease",inplace=True,axis=1)
```

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(0.95)
X_scaled_pca = pd.DataFrame(pca.fit_transform(X_scaled),columns=X_scaled_pca.columns)
```

```python
X_scaled_pca
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
```

```python
model1=RandomForestClassifier()
model2=SVC()
model3=LogisticRegression()
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X_scaled_pca,Y_scaled,test_size=0.2,random_state=42)
```

```python
model1.fit(X_train,Y_train)
```

```python
model2.fit(X_train,Y_train)
```

```python
model3.fit(X_train,Y_train)
```

```python
model1.score(X_test,Y_test)
```

```python
model2.score(X_test,Y_test)
```

```python
model3.score(X_test,Y_test)
```
